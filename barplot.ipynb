```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/data.csv', index_col='Responder_id')
df.head(3)
```

```python
from collections import Counter
```

```python
language_count = Counter()
for row in df['LanguagesWorkedWith']:
    language_count.update(row.split(';'))
    
print(language_count)
```

```python
df_lan = pd.DataFrame(language_count.most_common(10),columns=['language','counts'])
df_lan.set_index('language', inplace=True)
df_lan.head(2)
```

```python
df_lan.sort_values(by='counts').plot(kind='barh')
plt.title('most popular languages')
```

```python
languages = [x[0] for x in language_count.most_common(10)]
languages.reverse()
popularity = [x[1] for x in language_count.most_common(10)]
popularity.reverse()

plt.barh(languages, popularity)
```
