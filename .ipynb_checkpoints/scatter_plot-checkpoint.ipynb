> scatter plot is great to reveal relationship/ corelation

```python
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x, y);
```

```python
# style
plt.scatter(x, y, s=100, c='green', marker='x');
```

```python
plt.scatter(x, y, s=100, c='green', edgecolors='black', linewidth=1, alpha=.75);
```

```python
# represent the rating
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
```

```python
plt.scatter(x, y, c=colors, linewidth=1, edgecolor='black')
```

```python
plt.scatter(x, y, c=colors, cmap='Greens', linewidth=1, edgecolor='black')
```

```python
plt.scatter(x, y, c=colors, cmap='Greens', edgecolor='black', linewidth=1)
cbar = plt.colorbar()
cbar.set_label('satisfaction')
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black', linewidth=1)
cbar = plt.colorbar()
cbar.set_label('satisfaction')
```

# realworld data

```python
df = pd.read_csv('data/2019-05-31-data.csv')
df.head(3)
```

```python
plt.scatter(df['view_count'], df['likes'], edgecolor='black', linewidth=1, alpha=.75)
```

# find and remove the outlier

```python
mask = df['likes'] > 10**6
df.loc[mask]
```

```python
df = df[~mask]
plt.scatter(df['view_count'], df['likes'], edgecolor='black', linewidth=1, alpha=.75)
```

# also we can use log reprentation

```python
df = pd.read_csv('data/2019-05-31-data.csv')
df.head(3)
```

```python
plt.scatter(df['view_count'], df['likes'], edgecolor='black', linewidth=1, alpha=.75)
plt.xscale('log')
plt.yscale('log')
```

```python
plt.scatter(df['view_count'], df['likes'], c=df['ratio'], cmap='summer', 
            edgecolor='black', linewidth=1, alpha=.75)
cbar = plt.colorbar()
cbar.set_label('like/ dislike ratio')
plt.xscale('log')
plt.yscale('log')
```
