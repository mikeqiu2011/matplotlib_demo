```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xvals = np.arange(0, 10, 0.1)
plt.plot(xvals, np.sin(xvals))
```

```python
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['figure.dpi'] = 150
```

# labels

```python
plt.plot(xvals, np.sin(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

# multiple plot

```python
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
```

# line width

```python
plt.plot(xvals, np.sin(xvals), linewidth=3)
plt.plot(xvals, np.cos(xvals))
```

```python
plt.plot(xvals, np.cos(xvals))
plt.tight_layout()
```

```python
# grid
plt.plot(xvals, np.cos(xvals))
plt.grid(True)
```

# style

```python
plt.style.available
```

```python
plt.style.use('ggplot')
plt.plot(xvals, np.sin(xvals))
```

# legend and title

```python
plt.plot(xvals, np.sin(xvals), label=r'sin(x)')
plt.plot(xvals, np.cos(xvals), label=r'cos(x)')
plt.legend()
```

```python
# save
plt.plot(xvals, np.sin(xvals))
plt.savefig('plt.png')
```

# line color

```python
plt.plot(xvals, np.sin(xvals), label=r'sin(x)', color='red')
```

# line style

```python
plt.plot(xvals, np.sin(xvals), marker='o', linestyle='none')
plt.plot(xvals, np.cos(xvals), linestyle='--')
```

# Axis limit

```python
plt.plot(xvals, np.sin(xvals), color='orange', linestyle='--')
plt.ylim(0, 1)
```

```python
plt.plot(xvals, np.sin(xvals), color='orange', linestyle='--')
plt.xlim(0, 1)
```
