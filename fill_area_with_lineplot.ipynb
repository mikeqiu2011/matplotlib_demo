```python
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('data/salary.csv')
df.head(3)
```

```python
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']
```

```python
plt.plot(ages, dev_salaries, label='All devs', linestyle='--');
plt.plot(ages, py_salaries, label='Python devs')
# plt.plot(ages, js_salaries, label='JavaScript devs')
plt.legend();
```

```python
# fill area
plt.plot(ages, dev_salaries, label='All devs', linestyle='--', color='black', linewidth=1);
plt.plot(ages, py_salaries, label='Python devs')

plt.fill_between(ages, py_salaries, alpha=.25)

plt.legend();
```

```python
# fill area until customized medium
plt.plot(ages, dev_salaries, label='All devs', linestyle='--', color='black', linewidth=1);
plt.plot(ages, py_salaries, label='Python devs')

overall_median = 57287
plt.fill_between(ages, py_salaries, y2=overall_median, alpha=.25)

plt.legend();
```

```python
# fill area until customized medium
plt.plot(ages, dev_salaries, label='All devs', linestyle='--', color='black', linewidth=1);
plt.plot(ages, py_salaries, label='Python devs')

overall_median = 57287
plt.fill_between(ages, py_salaries, y2=overall_median, 
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=.25)

plt.legend();
```

```python
# also fill where is less than
plt.plot(ages, dev_salaries, label='All devs', linestyle='--', color='black', linewidth=1);
plt.plot(ages, py_salaries, label='Python devs')

overall_median = 57287
plt.fill_between(ages, py_salaries, y2=overall_median, 
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=.25)

plt.fill_between(ages, py_salaries, y2=overall_median, 
                 where=(py_salaries <= overall_median), 
                 interpolate=True, alpha=.25)

plt.legend();
```

```python
plt.plot(ages, dev_salaries, label='All devs', linestyle='--', color='black', linewidth=1);
plt.plot(ages, py_salaries, label='Python devs')

plt.fill_between(ages, py_salaries, y2=dev_salaries,
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha=.25, label='Above Avg')

plt.fill_between(ages, py_salaries, y2=dev_salaries,
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, alpha=.25, label='Below Avg')

plt.legend();
```
