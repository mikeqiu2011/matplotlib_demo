```python
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
ages = [18,19,21,25,26,26,30,32,38,45,55]
```

```python
# good for display distribution
plt.hist(ages, bins=5, edgecolor='black');
```

```python
# manually define bins, preferable way
bins= range(10, 70, 10)
plt.hist(ages, bins=bins, edgecolor='black');
```

```python
# remove under 20
bins= range(20, 70, 10)
plt.hist(ages, bins=bins, edgecolor='black');
```

```python
df = pd.read_csv('data/hist.csv')
df.head(3)
```

```python
ages = df['Age']
bins = range(10, 100, 10)
plt.hist(ages, bins=bins, edgecolor='black');
```

```python
plt.hist(ages, bins=bins, edgecolor='black', log=True);
```

```python
median_age = df['Age'].mean()
color= '#fc4f30'
plt.axvline(median_age, color=color, label='Age median')
plt.hist(ages, bins=bins, edgecolor='black');
plt.legend();
```
